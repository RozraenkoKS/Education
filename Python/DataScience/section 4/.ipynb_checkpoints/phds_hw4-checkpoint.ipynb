* монотонные функции от одного признака;
* немонотонные функции от одного признака;
* линейные функции от нескольких признаков;
* нелинейные функции от нескольких признаков;
* стандартизация признаков;
* преобразования категориальных признаков.

Например, вместо/дополнительно с признаком $x_j$ рассматривать признак $f(x_j)$, для указанных выше типов функций.

## 1. Монотонные функции от одного признака.
При замене признака $x_j$ на $f(x_j)$ соответствующим монотонным образом изменится порог для каждого объекта, соответствующего признакам. Остаётся только учесть область орпедления функции, чтобы не случилось выбросов. Структура дерева не изменится
## 2. Немонотонные функции от одного признака.
Здесь тоже могут случиться выбросы для функций с ограниченнной областью определения. Чётные функции повлияют на знак, и некоторые объекты станут относится к одному классу. Аналогично влияние других немонотонных функций. То есть поведение непредсказуемое, структура дерева изменится
## 3. Линейные функции от несокльких признаков.
При применении линейных преобразований мы, можно сказать, складываем и вычитаем признаки с весами. Но у разных признаков своя размерность. Допустим до преобразований 0-м узлом был x > 50. Пускай один признак имел значение 100, а другой 0.01. Применим линейную функцию и заменим 100 на 100.01, а 0.01 на 99.99. Значит, структура дерева изменится. Возможно, такие преобразования можно делать после стандартизации. Если имеется в виду линейное преобразование для каждого столбца признаков в отдельности, то так делать можно (см. п. 1)

Каждое наблюдение описывается 13 столбцами признаков и 1 столбцом класса, который идентифицирует его как кубическую, тетрагональную, ромбическую или ромбоэдрическую структуру.
```python
data = pd.read_csv("Crystal_structure.csv")
```

Сделаем предобработку данных.

```python
# уберем признаки, которые нам не нужны
data = data.drop(["In literature", "Compound", "τ"], axis=1)

# убираем объекты, для которых неизвестен таргет
data = data.drop(data[data["Lowest distortion"] == "-"].index).reset_index(drop=True)

# закодируем целевую переменную числами
data['Lowest distortion'] = LabelEncoder().fit_transform(data['Lowest distortion'])

# закодируем категориальные признаки числами
data['A'] = LabelEncoder().fit_transform(data['A'])
data['B'] = LabelEncoder().fit_transform(data['B'])

# Заполним пропуски нулями
data = data.replace('-', 0)
```

Подберите оптимальное значение максимальной глубины дерева. Для этого обучите решающие деревья с выставленным значением соответствующего гиперпараметра и посчитайте ошибку предсказаний каждого из них на валидационной выборке. Постройте соответствующий график - зависимость ошибки от максимальной глубины дерева. Подробнее про гиперпараметры смотрите в [документации](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html).

```python
depths_r2_scores = []
depths = range(1, 20)
for depth in depths:
    model = DecisionTreeRegressor(max_depth=depth)
    model.fit(X_train, y_train)
    depths_r2_scores.append(r2_score(y_valid, model.predict(X_valid)))

plt.figure(figsize=(8, 6))
plt.scatter(depths, depths_r2_scores)
plt.xticks(depths[::2])
plt.xlabel('Глубина')
plt.ylabel('R2')
plt.title('Зависимость метрики R2 от гиперпараметра максимальной глубины дерева')
plt.plot()
```

```python
opt_depth = 12
```

Аналогичным образом подберите минимальное число объектов, необходимое для того, чтобы узел дерева мог бы расщепиться, зафиксировав при этом подобранное значение максимальной глубины. Для минимального числа объектов в листе и максимального числа признаков, по которым ищется лучшее разбиение в дереве, подберите аналогично, фиксируя лучшие значения гиперпараметров, подобранные на предыдущем шаге.

```python
min_samples_splits_r2_scores = []
min_samples_splits = range(2, 10)
for min_samples_split in min_samples_splits:
    model = DecisionTreeRegressor(max_depth=opt_depth, min_samples_split=min_samples_split)
    model.fit(X_train, y_train)
    min_samples_splits_r2_scores.append(r2_score(y_valid, model.predict(X_valid)))

plt.figure(figsize=(8, 6))
plt.scatter(min_samples_splits, min_samples_splits_r2_scores)
plt.xlabel('Минимальное число для расщипления')
plt.ylabel('R2')
plt.title('Зависимость метрики R2 от гиперпараметра минимального числа для расщипления')
plt.plot()
```

```python
opt_min_samples_split = 2
```

```python
min_samples_leafs_r2_scores = []
min_samples_leafs = range(1, 10)
for min_samples_leaf in min_samples_leafs:
    model = DecisionTreeRegressor(max_depth=opt_depth, 
                                  min_samples_split=opt_min_samples_split, 
                                  min_samples_leaf=min_samples_leaf)
    model.fit(X_train, y_train)
    min_samples_leafs_r2_scores.append(r2_score(y_valid, model.predict(X_valid)))

plt.figure(figsize=(8, 6))
plt.scatter(min_samples_leafs, min_samples_leafs_r2_scores)
plt.xlabel('Минимальное число для листа')
plt.ylabel('R2')
plt.title('Зависимость метрики R2 от гиперпараметра минимального числа для расщипления')
plt.plot()
```

```python
opt_min_samples_leaf = 1
```

```python
max_featuress_r2_scores = []
max_featuress = range(1, 10)
for max_features in max_featuress:
    model = DecisionTreeRegressor(max_depth=opt_depth, 
                                  min_samples_split=opt_min_samples_split, 
                                  min_samples_leaf=min_samples_leaf,
                                  max_features=max_features)
    model.fit(X_train, y_train)
    max_featuress_r2_scores.append(r2_score(y_valid, model.predict(X_valid)))


max_featuress_mod = [None, 'sqrt', 'log2']
for max_features in max_featuress_mod:
    model = DecisionTreeRegressor(max_depth=opt_depth, 
                                      min_samples_split=opt_min_samples_split, 
                                      min_samples_leaf=min_samples_leaf,
                                      max_features=max_features)
    model.fit(X_train, y_train)
    mod_max_feature_r2_score = r2_score(y_valid, model.predict(X_valid)) 
    print(f'При max_features = {max_features}, R2 score = {mod_max_feature_r2_score}')

plt.figure(figsize=(8, 6))
plt.scatter( max_featuress, max_featuress_r2_scores)
plt.xlabel('Максимально число учитываемых признаков')
plt.ylabel('R2')
plt.title('Зависимость метрики R2 от гиперпараметра минимального числа для расщипления')
plt.plot()
```
