### Задача 1

В этой задаче вам предстоит визуализировать центральную предельную теорему.

*a).* Пусть $\xi_1, \xi_2,..., \xi_n$ — независимые случайные величины, имеющие распределение $Exp(\lambda)$. Согласно центральной предельной теореме, имеет место сходимость $$Z_n = \frac{X_n - \mathsf{E}X_n}{\sqrt{\mathsf{D}X_n}} \stackrel{d}{\longrightarrow} \mathcal{N}(0, 1),$$ где $X_n = \sum_{i=1}^n \xi_i$.

Чтобы убедиться в этом, вам необходимо сгенерировать множество наборов случайных величин и рассчитать значение $Z_n$ для каждого набора, в зависимости от его размера.

Сгенерируйте 1000 наборов случайных величин $\xi_1^j,..., \xi_{300}^j$ из распределения $Exp(1)$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
size = 300 # размер выборки
samples_count = 1000  # количество выборок

rng = np.random.default_rng()
scale = 1
sample = rng.exponential(size=(samples_count, size), scale=scale)
```

По каждому из них посчитайте сумму $X_{jn} = \sum\limits_{i=1}^n \xi_i^j$ для $1 \leqslant n \leqslant 300$, то есть сумма первых $n$ величин $j$-го набора.

```python
sums = np.cumsum(sample, axis=1)
```

Для этого среднего посчитайте величину $Z_{jn} = \frac{X_{jn} - \mathsf{E}X_{jn}}{\sqrt{\mathsf{D}X_{jn}}}$.

*Подсказка: можно воспользоваться возможностями библиотеки `numpy`.*

```python
n_values = np.arange(1, size + 1)
expected = n_values  # Математическое ожидание
std_dev = np.sqrt(n_values)  # Стандартное отклонение
```

```python
z = (sums - expected) / std_dev
```

Для каждого $j$ нанесите на один график зависимость $Z_{jn}$ от $n$ с помощью `plt.plot`. Каждая кривая должна быть нарисована *одним цветом* с прозрачностью `alpha=0.05`. Сходятся ли значения $Z_{jn}$ к какой-либо константе?

```python
plt.figure(figsize=(8, 6))
plt.plot(n_values, z.T, alpha=0.05, color='b')  
plt.axhline(0, color='r', alpha=0.6)  
plt.xlabel("n")
plt.ylabel("$Z_{jn}$")
plt.title("Центральная предельная теорема для Exp(1)")
plt.show()
```


Для $n=300$ по набору случайных величин $Z_{1,300}, ..., Z_{1000,300}$ постройте гистограмму. Похожа ли она на плотность распределения $\mathcal{N}(0, 1)$, которую тоже постройте на том же графике? Не забудьте сделать легенду (подписи на графиках, см. семинар и туториалы).

```python
plt.hist(z[:, -1], bins=30, density=True, alpha=0.7, label='Гистограмма $Z_{jn}$')
plt.hist(rng.normal(loc=0, scale=1, size=samples_count), 
         bins=30, density=True, alpha=0.5, label='Гистограмма нормального распределения')
plt.title(r'Сравнение распределений $Z_{jn}$ и $\mathcal{N}(0,1)$')  # Добавлен raw-строка r''
plt.legend()
plt.show()
```

*b).* Выполните те же действия для распределения $Pois(1)$.


```python
size = 300 # размер выборки
samples_count = 1000  # количество выборок

rng = np.random.default_rng()
lam = 1
sample = rng.poisson(size=(samples_count, size), lam=lam)

sums = np.cumsum(sample, axis=1)

n_values = np.arange(1, size + 1)
expected = n_values * lam  # Математическое ожидание
std_dev = np.sqrt(n_values * lam)  # Стандартное отклонение

z = (sums - expected)/(std_dev)

plt.figure(figsize=(8, 6))
plt.plot(n_values, z.T, alpha=0.05, color='b')
plt.axhline(0, color='r', alpha = 0.6) 
plt.xlabel("n")
plt.ylabel("$Z_{jn}$")
plt.title("Центральная предельная теорема для Poisson(1)")
plt.show()

# Гистограмма последних значений Z_{jn} (при n=300)
plt.hist(z[:, -1], bins=30, density=True, alpha=0.7, label='Гистограмма $Z_{jn}$')
plt.hist(rng.normal(loc=0, scale=1, size=samples_count), 
         bins=30, density=True, alpha=0.5, label='Гистограмма нормального распределения')

plt.title(r'Сравнение распределений $Z_{jn}$ и $\mathcal{N}(0,1)$')
plt.legend()
plt.show()
```

<font color="green">**Профиль биология**</font>

Время жизни лейкоцита подчиняется показательному распределению с математическим ожиданием 200 часов. Найти вероятность того, что среднее время жизни выборки из 100 лейкоцитов окажется не менее 180 часов.

```python
import scipy.stats as stats
```

```python
# Заданные параметры:
mu = 100        # математическое ожидание (микросекунды)
sigma = 100     # стандартное отклонение экспоненциального распределения (также 100 мкс)
n = 100         # размер выборки

# Стандартная ошибка среднего:
std_error = sigma / np.sqrt(n)  # = 100 / 10 = 10

# Пороговое значение для среднего:
x_bar_threshold = 90

# Вычисляем z-статистику:
z = (x_bar_threshold - mu) / std_error  # z = (90 - 100) / 10 = -1

# По центральной предельной теореме, выборочное среднее распределено примерно нормально,
# т.е. P( x̄ >= 90 ) = P(Z >= -1).
# Используя симметрию нормального распределения:
# P(Z >= -1) = P(Z <= 1) = F(1).

# Вычисляем вероятность:
probability = stats.norm.cdf(1)

print("Вероятность, что среднее время когерентности выборки из 100 кубитов ≥ 90 мкс:",
      probability)
```

# Дерево 2

## Описание дерева

- **Единый узел (лист):**  
  - **Условие:** Если одновременно выполняются все следующие условия:  
    - Признак **Su**: 340 < Su < 550,  
    - Признак **Sy**: 210 < Sy < 420,  
    - Признак **E**: 1 680 000 < E < 2 180 000,  
    - Признак **G**: 55 000 < G < 95 000,  
    - Признак **Ro**: 7100 < Ro < 7900,  
    - Признак **mu** принимает значение из набора {0.27, 0.29, 0.30}.  
  - **Действие:**  
    - Если условие выполнено – классифицируем объект как *Use: True*.  
    - Иначе – классифицируем как *Use: False*.

## Пояснение

*Объединённое правило, содержащее проверку всех необходимых признаков в одном узле, позволяет компактно и эффективно определить истинные объекты (Use: True). Такой подход исключает промежуточные ветвления, что упрощает модель, сохраняя при этом достаточную точность классификациию*

Реализуйте полученные деревья по шаблону

```python
def tree(features: pd.DataFrame) -> pd.Series:
    """
    Предсказание целевого признака для данных features -- pandas-таблица данных.
    Возвращает pandas.Series с теми же индексами, что и у features.
    """
    
    X_1 = ((features['Su'] > 340) & (features['Su'] < 550) &
           (features['Sy'] > 210) & (features['Sy'] < 420))
    X_2 = ((features['E'] > 168000) & (features['E'] < 218000) &
           (features['G'] > 55000) & (features['G'] < 95000) &
           (features['Ro'] > 7100) & (features['Ro'] < 7900))

    predicted = X_1 & X_2
    
    return predicted
```

&#x2753; **Вопрос** &#x2753;
> Как это может повлиять на работу kNN, почему?

**Ваш ответ:** Если разброс большой, то разные классы имею разный вес расстояний до их членов. Допустим если разброс малый, то расстояний до этого класса будет относительно небольшим. А если разброс большой, то расстояния могут быть на порядки выше, чем в предыдщуем случае. Но в одном масштабе эти расстояния могут быть сравнимы


Для решения этой проблемы нужно привести признаки к единому масштабу. Существует много методов масштабирования рассмотрим работу одного из них &mdash; **минимаксная нормализация**

  
> **Минимакс** &mdash; линейное преобразование данных в диапазоне $[0, 1]$, где минимальное и максимальное масштабируемые значения соответствуют 0 и 1 соответственно.
> Реализуется оно по следующей формуле:
>
> $$\widetilde{x}_{ij} = \frac{x_{ij} - m_j}{M_j - m_j}$$
>
> где $x_{ij}$ &mdash; значение признака $j$ для объекта $i$, а $m_j$ и $M_j$ &mdash; минимальное и максимальное значения признака $j$ в обучающем наборе данных.  
> В модуле `sklearn.preprocessing` есть [класс](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html) `MinMaxScaler`, который реализует эту операцию. Давайте рассмотрим методы этого класса подробнее:
>
> * `fit(X)`: вычисляет минимальные и максимальные значения для каждого признака.
>
> * `transform(X)`: применяет масштабирование к данным.
>
> * `fit_transform(X)`: объединяет оба метода, сначала подгоняя модель, а затем трансформируя данные.
>
> * `inverse_transform(X)`: восстанавливает исходные данные из масштабированных.


Стоит отметить что существуют и другие методы масштабирования данных.
