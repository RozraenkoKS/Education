#### 2. Предобработка данных

В этом разделе ожидается, что вы обработаете пропуски (если они есть) с помощью линейной регрессии, а выбросы (если они есть) обработайте способом на ваше усмотрение. Для оценки качества обработки пропусков используйте метрику MAPE.

```python
l1pt = X['L1pT']
mtr2 = X['M_TR_2']
X_dropped = X.drop(['L1pT', 'M_TR_2'], axis=1)

def regression_impute(features, target):
    """
    Функция заменяет пропущенные значения в target, используя линейную регрессию
    с данными features в качестве признаков.
    """
    null = target.isnull()
    notnull = ~null
    
    if null.sum() == 0:
        return target
    
    y_train = target[notnull]
    X_train = features[notnull]
    X_missing = features[null]

    model = LinearRegression()
    model.fit(X_train, y_train)
    pred = model.predict(X_missing)
    
    target.loc[null] = pred
    return target

for col in ['L1pT', 'M_TR_2']:
    X[col] = regression_impute(X_dropped, X[col])

    
                   
```

### Задача 2

В этой задаче вам нужно будет реализовать оптимальный подсчет метрики **ROC-AUC**. Вспомним, что **ROC** &mdash; кривая в координатах **True Positive Rate (TPR)** и **False Positive Rate (FPR)**, а **ROC-AUC** &mdash; площадь под этой кривой.

**TPR (True Positive Rate, полнота)** показывает какую долю объектов положительного класса модель классифицировала правильно:

$$TPR = \dfrac{TP}{TP + FN}$$


**FPR (False Positive Rate)** показывает, какую долю из объектов негативного класса модель предсказала неверно:

$$FPR = \dfrac{FP}{FP + TN}$$

1. Предложите алгоритм расчета метрики за $\mathcal{O} \left(n \log n\right)$, где $n$ &mdash; размер выборки.  
    *Подсказка.* Подумайте сколько всего различных пар $\left(FPR, TPR\right)$ существует.

    Пускай m - число единиц, а n - число 0. Тогда разбив прямоугольник (0; 0), (0; 1), (1; 1), (1; 0), смыслу ROC кривой будет соответствовать следующее построение: стартуя из точки (0; 0), мы, начиная с самой высокой предсказанной вероятности (и вниз) в каждом узле проводим прямую вверх, если в соотвествующей точке предсказан класс 1, вправо, если 0. Если предсказания равны, то делаем шаг на a вверх и на b вправо, где a - число 1, а b - число 0 с данной оценкой. Таким образом получим m * n пар TPR и FPR. 

2. Реализуйте предложенный алгоритм.

```python
def roc_auc(y_true, y_score):
    '''
    Вычисляет ROC-AUC за O(n log(n))
    :param y_true: истинные бинарные метки
    :param y_score: предсказанная вероятность положительного класса
    '''

    # '''
    # Вычисляет ROC-AUC за O(n ^ 2)
    # :param y_true: истинные бинарные метки
    # :param y_score: предсказанная вероятность положительного класса
    # '''
    # n = len(y_true)
    # auc_num = 0
    # auc_den = 0

    # def I(y_i, y_j):
    #     if y_i < y_j:
    #         return 1
    #     return 0

    # def I_(a_i, a_j):
    #     if a_i < a_j:
    #         return 1
    #     elif a_i == a_j:
    #         return 0.5
    #     return 0
        
    # for i in range(n):
    #     for j in range(n):
    #         x = I(y_true[i], y_true[j])
    #         auc_den += x
    #         auc_num += x * I_(y_score[i], y_score[j])
    # return auc_num / auc_den

    y_true = np.array(y_true)
    y_score = np.array(y_score)
    

    m = np.sum(y_true)             
    n = len(y_true) - m          
    
    if m == 0 or n == 0:
        return 0.5

    order = np.argsort(-y_score)
    y_true_sorted = y_true[order]
    y_score_sorted = y_score[order]

    groups = []
    i = 0
    total = len(y_true_sorted)
    while i < total:
        current_score = y_score_sorted[i]
        group = []
        while i < total and y_score_sorted[i] == current_score:
            group.append(y_true_sorted[i])
            i += 1
        groups.append(group)

    current_x = 0.0  
    current_y = 0.0  
    auc = 0.0

    for group in groups:

        a = sum(group)      
        b = len(group) - a     

        new_x = current_x + (b / n)
        new_y = current_y + (a / m)

        delta_x = new_x - current_x
        auc += (current_y + new_y) / 2 * delta_x

        current_x, current_y = new_x, new_y
        
    return auc
```
