### Задача 3


Пусть $X_1, ..., X_n$ &mdash; выборка из распределения $\mathcal{N}(0, \sigma^2)$. Известно, что в качестве оценки параметра $\sigma^2$ можно использовать следующие оценки $S^2, \frac{n}{n-1}S^2$. Проверьте эти оценки на несмещенность, повторив эксперимент с семинара.

Реализуйте две функции, каждая из которых на вход берет несколько выборок, а на выход выдает массив оценок (первая функция - $S^2$, вторая функция - $\frac{n}{n-1}S^2$)  для каждой выборки.  
*Примечание:* У матрицы, заданной в виде `numpy.array` есть метод `var` вычисляющий выборочную дисперсию.

```python
def estimate_s2(X):
    '''
    Принимает на вход массив размером (n_samples, sample_size), выдает массив оценок размера (n_samples,)
    '''
    return np.var(X, axis=1)

def estimate_s2_corrected(X):
    '''
    Принимает на вход массив размером (n_samples, sample_size), выдает массив оценок размера (n_samples,)
    '''
    n = X.shape[1]
    return np.var(X, axis=1) * n / (n - 1)
```

Проверьте правильность своей реализации

```python
X = np.array([[1, 3, 3, 6], [2, 5, 1, 89]])
estimate_s2(X)

assert(sum(estimate_s2(X) != np.array([3.1875, 1399.6875])) == 0)
assert(sum(estimate_s2_corrected(X) != np.array((4.25, 1866.25))) == 0)
```

```python
estimators = [
    (estimate_s2, r'$S^2$', 'purple', 0.1),
    (estimate_s2_corrected, r'$\frac{n}{n-1}S^2$', 'green', 0.2)
]
```

Мы хотим понять, являются ли оценки параметра $\theta$ смещенными, для этого нам нужно провести множество экспериментов (сгенерировать выборку много раз, так как при каждой генерации получаются разные числа).

Напишите функцию построения таких графиков в общем виде: на вход функция может получать любые распределения (`distributions`) и любые оценки (`estimators`). 
Средние значения по экспериментам для каждой оценки нужно обозначить звездочками.

*Подсказка: получается, что для этой задачи достаточно просто взять функции из семинарского ноутбука?!*

```python
def est_plot(distribution, estimators, sample_size, sample_count):
    '''
    Построение графика разброса реализаций оценок и их средних значений.

    distribution -- распределение формата scipy.stats
    estimators -- список оценок и параметров для отрисовки графиков
    sample_size -- размер выборок
    sample_count -- количество генерируемых выборок
    '''
    X = distribution.rvs(size=(sample_count, sample_size))
    plt.figure(figsize=(15, len(estimators)))
    for estimator, name, color, y in estimators:
        E = estimator(X)
        plt.scatter(E, np.zeros(sample_count) + y, alpha=0.1,
                        s=100, color=color, label=name)
        plt.scatter(E.mean(), y, marker='*', s=300,
                        color='w', edgecolors='black')

        plt.vlines(1, -1, 1, color='r')
        plt.title('Размер выборки = %d' % sample_size)
        plt.yticks([])
        plt.legend()
        plt.xlim((0.8, 1.2))
        plt.ylim((-0.1, 0.2 * len(estimators)))
```

Постройте три графика для разных размеров выборок (10, 100, 500).


```python
sample_size_list = (10, 100, 500)  # размеры выборок
sample_count = 500  # количество экспериментов

for sample_size in sample_size_list:
    est_plot(distribution=sps.norm, estimators=estimators, sample_size=sample_size, sample_count=sample_count)
plt.show()
```

Пусть $X_1, X_2, \dots, X_n$ — независимые одинаково распределённые случайные величины с экспоненциальным распределением, имеющим функцию плотности

$$
f(x;\lambda) = \lambda \, e^{-\lambda x}, \quad x \ge 0, \quad \lambda > 0.
$$

**Шаг 1.** Запишем функцию правдоподобия для выборки:

$$
L(\lambda) = \prod_{i=1}^n \lambda \, e^{-\lambda x_i} = \lambda^n \, e^{-\lambda \sum_{i=1}^n x_i}.
$$

**Шаг 2.** Найдём логарифм функции правдоподобия:

$$
\ell(\lambda) = \ln L(\lambda) = n \ln \lambda - \lambda \sum_{i=1}^n x_i.
$$

**Шаг 3.** Для нахождения оценки максимального правдоподобия (ОМП) найдём производную логарифма правдоподобия по $\lambda$ и приравняем её к нулю:

$$
\frac{d\ell}{d\lambda} = \frac{n}{\lambda} - \sum_{i=1}^n x_i = 0.
$$

**Шаг 4.** Решим уравнение относительно $\lambda$:

$$
\frac{n}{\lambda} = \sum_{i=1}^n x_i \quad \Rightarrow \quad \lambda = \frac{n}{\sum_{i=1}^n x_i}.
$$

Таким образом, оценка максимального правдоподобия для параметра $\lambda$ имеет вид

$$
\hat{\lambda} = \frac{n}{\sum_{i=1}^n x_i},
$$

что эквивалентно $\hat{\lambda} = \frac{1}{\overline{x}}$, где $\overline{x}$ — выборочное среднее.

Это и есть искомая ОМП для параметра экспоненциального распределения.

Рассмотрим модель линейной регрессии

$$Y = X \theta + \varepsilon,$$

где $Y \in \mathbb{R}^n$ - отклик, $X \in \mathbb{R}^{n \times d}$, $\theta \in \mathbb{R}^d$, $\varepsilon \in \mathbb{R}^n$ - шум, $n > d$, $rk X = d$.

  $$\widehat{\theta} = (X^TX)^{-1}X^T Y$$

  является несмещенной оценкой параметра $\theta$. Будет ли выполнено свойство несмещенности для Ridge-регрессии?

  *Указание:* используйте свойство линейности математического ожидания в матричном виде.
  
3. Покажите, что, если дополнительно матрица ковариаций $D \varepsilon = \sigma^2 I_n$, дисперсия МНК-оценки $D \widehat{\theta} = \sigma^2 (X^TX)^{-1}$. Что показывает данный результат в случае мультиколлинеарности признаков?
  *Указание:* используйте свойство дисперсии в матричном виде.


**Напоминание**

1. Матричное свойство математического ожидания:

  $∀ A \in \mathbb{R}^{m \times k}$, случайного вектора $\psi \in \mathbb{R}^k$

  $$E A \psi = A E \psi$$
2. Матричное свойство дисперсии:
  $∀ A \in \mathbb{R}^{m \times k}$, случайного вектора $\psi \in \mathbb{R}^k$

  $$D A \psi = A (D \psi) A^T$$

1. Случаной является вектор $\varepsilon$, матрица $Y$. Неизвестным - $\theta$
2. $$
\widehat{\theta} = (X^T X)^{-1} X^T Y
$$
$$E \widehat{\theta} = (X^TX)^{-1}X^T Y = E (X^TX)^{-1}X^T (X \theta + \varepsilon) = \theta$$

то есть является несмещённой. Ridge-регрессия с оценкой

$$
\widehat{\theta}_{\text{ridge}} = (X^T X + \lambda I)^{-1} X^T Y
$$

$$
E[\widehat{\theta}_{\text{ridge}}] = (X^T X + \lambda I)^{-1} X^T X\,\theta \neq \theta,
$$

то есть является смещённой.

3. Пусть дана модель
$$
Y = X \theta + \varepsilon,
$$
где шум $ \varepsilon \in \mathbb{R}^n $ имеет матрицу ковариаций
$$
D \varepsilon = \sigma^2 I_n.
$$

МНК-оценка определяется как
$$
\widehat{\theta} = (X^T X)^{-1} X^T Y.
$$

Подставляем модель:
$$
\widehat{\theta} = (X^T X)^{-1} X^T (X \theta + \varepsilon)
= \theta + (X^T X)^{-1} X^T \varepsilon.
$$

Чтобы найти дисперсию $\widehat{\theta}$, воспользуемся матричным свойством:
$$
D(A\psi) = A\,(D\psi)\,A^T,
$$
где $A$ — любая матрица, а $\psi$ — случайный вектор.

Применяем это свойство с $A = (X^T X)^{-1} X^T$ и $\psi = \varepsilon$:
$$
D\widehat{\theta} = (X^T X)^{-1} X^T\,(D\varepsilon)\,X\,(X^T X)^{-1}.
$$

Так как $D\varepsilon = \sigma^2 I_n$, получаем:
$$
D\widehat{\theta} = \sigma^2\,(X^T X)^{-1} X^T X\,(X^T X)^{-1}.
$$

Заметим, что $X^T X\,(X^T X)^{-1} = I_d$, поэтому:
$$
D\widehat{\theta} = \sigma^2 (X^T X)^{-1}.
$$

#### Интерпретация при мультиколлинеарности

При мультиколлинеарности признаки в матрице $X$ сильно коррелируют друг с другом, что приводит к почти вырождённости матрицы $X^T X$. В этом случае её обратная матрица $(X^T X)^{-1}$ имеет очень большие элементы, и, как следствие, дисперсия МНК-оценки
$$
D\widehat{\theta} = \sigma^2 (X^T X)^{-1}
$$
становится очень большой.

Это означает, что оценки параметров $\theta$ становятся нестабильными и имеют высокую изменчивость. Даже небольшие изменения в данных могут привести к существенным колебаниям оценок, что затрудняет их интерпретацию и надежное использование модели.

  - Задайте сами истинное значение $\theta$ и значения признаков $X \in \mathcal{R}^{n \times d}$. Учитывайте, что для выполнения условия теоремы с лекции $n > d$ и $rk X = d$. Для Ridge-регрессии возьмите два значения $\lambda$. Сгенерируйте зашумленные значения $Y$, с помощью которых можно получить выборки для каждого вида оценки. Визуализируйте их аналогично задаче 3. Распределение шума $\varepsilon$ можно считать нормальным. 
  
  **Внимание:** вы должны получить $m$ реализаций $Y$, по которым будут посчитаны $m$ реализаций оценок каждого вида.

Зафикисируем в нормальном распрделении шума матожидание 0, матрицу Х и истинные значения вектора параметров.

$Y = X \theta + \varepsilon$

где $Y \in \mathbb{R}^n$ - отклик, $X \in \mathbb{R}^{n \times d}$, $\theta \in \mathbb{R}^d$, $\varepsilon \in \mathbb{R}^n$ - шум, $n > d$, $rk X = d$.

$\widehat{\theta}_{МНК} = (X^TX)^{-1}X^T Y$

$\widehat{\theta}_{Ridge} = (X^T X + \lambda I)^{-1} X^T Y$

```python
n = 100
d = 50
m = 10000

lambda1 = 2
lambda2 = 10

theta = sps.norm(scale=1).rvs(size=d)
while True:
    X = sps.norm(scale=1).rvs(size=(n, d))
    if np.linalg.matrix_rank(X) == d:
        break
```

```python
mnk = []
ridge1 = []
ridge2 = []

XtX = X.T @ X
X_pinv = np.linalg.pinv(X) 

for i in range(m):

    eps = sps.norm(scale=1).rvs(size=n)
    Y = X @ theta + eps

    theta_mnk = X_pinv @ Y

    theta_ridge1 = np.linalg.solve(XtX + lambda1 * np.eye(d), X.T @ Y)
    theta_ridge2 = np.linalg.solve(XtX + lambda2 * np.eye(d), X.T @ Y)

    mnk.append(theta_mnk)
    ridge1.append(theta_ridge1)
    ridge2.append(theta_ridge2)
```

```python
mnk = np.array(mnk)
ridge1 = np.array(ridge1)
ridge2 = np.array(ridge2)
```

```python
ncols = 3
x = np.arange(1, m+1)

fig, ax = plt.subplots(ncols=ncols, figsize=(15, 7))
for i in range(ncols):
    y_mnk = np.cumsum(mnk[:, i]) / x
    y_ridge1 = np.cumsum(ridge1[:, i]) / x
    y_ridge2 = np.cumsum(ridge2[:, i]) / x
    
    ax[i].plot(x, y_mnk, label=r'$\widehat{\theta}_{МНК}$')
    ax[i].plot(x, y_ridge1, label=r'$\widehat{\theta}_{Ridge, (\lambda = 2)}$')
    ax[i].plot(x, y_ridge2, label = r'$\widehat{\theta}_{Ridge, (\lambda = 10)}$')
    ax[i].axhline(theta[i], color='black', linestyle='--', label=r'Истинный $\theta$')
    ax[i].set_xlabel('Номер итерации')
    ax[i].set_ylabel('Среднее значение')
    ax[i].set_title(f'Компонента θ[{i}]')
    ax[i].legend()

plt.tight_layout()
plt.show()
```
